```python
#Importing necessary liberaries 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import IsolationForest
from keras.models import Sequential
from keras.layers import LSTM, Dense, Input
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
df = pd.read_csv('ambient_temperature_system_failure.csv')
df.head(10)
```

```python
df.ffill()
```

```python
# Data Normalization
scaler = MinMaxScaler()
df['value_scaled'] = scaler.fit_transform(df[['value']])
```

```python
#  Rolling Window Features
# Rolling mean and standard deviation
df['rolling_mean'] = df['value_scaled'].rolling(window=24).mean()
df['rolling_std'] = df['value_scaled'].rolling(window=24).std()
```

```python
#Exploratory Data Analysis (EDA)

#Visualizing the Time Series
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['value'], label='Original Data')
plt.plot(df.index, df['rolling_mean'], label='Rolling Mean', color='orange')
plt.fill_between(df.index, df['rolling_mean'] - 2 * df['rolling_std'], df['rolling_mean'] + 2 * df['rolling_std'], color='gray', alpha=0.2)
plt.legend()
plt.show()
```

```python
# Summary Statistics
df.describe()
```

```python
# MODEL SECTION 
# Initialize and train Isolation Forest
iso_forest = IsolationForest(contamination=0.01)
df['anomaly_score'] = iso_forest.fit_predict(df[['value_scaled']])
```

```python
# Plot anomalies
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['value'], label='Original Data')
anomalies = df[df['anomaly_score'] == -1]
plt.scatter(anomalies.index, anomalies['value'], color='red', label='Anomaly')
plt.legend()
plt.show()
```

```python
#  LSTM Model (Deep Learning)
# Prepare dataset for LSTM (reshaping data)
n_steps = 30  # Number of previous time steps
X, y = [], []
for i in range(n_steps, len(df)):
    X.append(df['value_scaled'].iloc[i-n_steps:i].values)
    y.append(df['value_scaled'].iloc[i])
```

```python
# Split data into train/test sets
X_train, y_train = X[:int(len(X)*0.8)], y[:int(len(y)*0.8)]
X_test, y_test = X[int(len(X)*0.8):], y[int(len(y)*0.8):]
```

```python
# Build LSTM Model
model = Sequential()
model.add(Input(shape=(n_steps, 1)))  # Explicitly define the input layer
model.add(LSTM(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
# Train the model
model.fit(np.array(X_train), np.array(y_train), epochs=10, batch_size=32, verbose=0)
```

```python
# Model Evaluation

# Predict and Evaluate
y_pred = model.predict(np.array(X_test))
```

```python
# Evaluate errors
mse = np.mean(np.square(np.array(y_test) - y_pred))
```

```python
# Define a threshold for anomalies
threshold = 0.01

# Ensure y_pred is flattened
y_pred_flat = y_pred.flatten()

# Calculate the error (difference between actual and predicted values)
error = abs(np.array(y_test) - y_pred_flat)

# Ensure the length of 'error' matches the test set
df_test = df.iloc[len(X_train):len(X_train) + len(y_test)]

# Define anomalies as points where the error exceeds the threshold
anomalies = df_test[error > threshold]

# Output the detected anomalies
print(anomalies)
```

```python
# Create binary arrays for true and predicted anomalies
y_true = np.where(abs(np.array(y_test) - np.mean(y_test)) > threshold, 1, 0)  # 1 if true anomaly, 0 otherwise
y_pred_binary = np.where(abs(np.array(y_pred) - np.mean(y_test)) > threshold, 1, 0)  # 1 if predicted as anomaly, 0 otherwise

# Calculate precision, recall, and F1 score
precision = precision_score(y_true, y_pred_binary)
recall = recall_score(y_true, y_pred_binary)
f1 = f1_score(y_true, y_pred_binary)

# Output the results
print(f"Precision: {precision}, Recall: {recall}, F1 Score: {f1}")
```

```python
# Save the model
model.save('lstrm_model.h5')  # Saves the model to a file
print("Model saved as lstm_model.h5")
```
